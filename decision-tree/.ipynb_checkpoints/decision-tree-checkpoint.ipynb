# Decision Tree

## Import tools

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pymysql

from math import sqrt
from sqlalchemy import create_engine

from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

pymysql.install_as_MySQLdb()
```

## Get the data

```python
engine = create_engine('mysql://root:0201@localhost:3306/household')
```

```python
with engine.begin() as connection:
    df = pd.read_sql("select * from household.clean_dataset", con=connection, index_col='dt')

df.head()
```

```python
len(df)
```

## Train-Test split

```python
X = df.drop("Sub_metering_3", axis=1)
y = df["Sub_metering_3"]
```

```python
np.random.seed(3005)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=41)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_test_tmp = X_test
```

```python
X_train = X_train.values
y_train = y_train.values.reshape(-1,1)
X_test = X_test.values
y_test = y_test.values.reshape(-1,1)
```

## Sklearn

```python
max_depth_list = [x for x in range(1,50,1)]
rmse_list = []

for i in max_depth_list:
    regressor = DecisionTreeRegressor(max_depth=i)
    regressor.fit(X_train, y_train)

    y_pred = regressor.predict(X_test)

    rmse = sqrt(mean_squared_error(y_test, y_pred))
    rmse_list.append(rmse)
```

```python
plt.plot(max_depth_list, rmse_list)
plt.xlabel("max_dept values")
plt.ylabel("RMSE")
```

### Tìm được max_depth =  12 (max_depth = index of min value + 1)

```python
max_depth = rmse_list.index(min(rmse_list)) + 1
max_depth
```

```python
regressor = DecisionTreeRegressor(max_depth=max_depth)
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test) 

np.sqrt(mean_squared_error(y_test, y_pred))
```

## Node class

```python
class Node():
    def __init__(self, feature_index=None, threshold=None, left=None, right=None, var_red=None, value=None):
        ''' constructor ''' 
        
        # for decision node
        self.feature_index = feature_index
        self.threshold = threshold
        self.left = left
        self.right = right
        self.var_red = var_red
        
        # for leaf node
        self.value = value
```

## Tree class

```python
class DecisionTreeRegressor():
    def __init__(self, min_samples_split=2, max_depth=2):
        ''' constructor '''
        
        # Khởi tạo gốc của cây (root được khởi tạo khi fit model)
        self.root = None
        
        # Điều kiện dừng
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
        
    def make_tree(self, dataset, curr_depth=0):
        ''' Hàm đệ quy để xây dựng cây quyết định '''
        
        X, Y = dataset[:,:-1], dataset[:,-1]
        num_samples, num_features = np.shape(X)
        best_split = {}
        # Tiếp tục xây dựng cây tới khi gặp điều kiện dừng
        if num_samples >= self.min_samples_split and curr_depth <= self.max_depth:
            # Tìm cách chia nhánh tốt nhất
            best_split = self.get_best_split(dataset, num_samples, num_features)
            # check if information gain is positive
            if best_split["var_red"]>0:
                # Nhánh trái
                left_subtree = self.make_tree(best_split["dataset_left"], curr_depth + 1)
                # Nhánh phải
                right_subtree = self.make_tree(best_split["dataset_right"], curr_depth + 1)
                
                return Node(best_split["feature_index"], best_split["threshold"], 
                            left_subtree, right_subtree, best_split["var_red"])
        
        # Tính giá trị tại nút lá
        leaf_value = np.mean(Y)
        
        return Node(value=leaf_value)
    
    def get_best_split(self, dataset, num_samples, num_features):
        ''' function to find the best split '''
        
        # dictionary to store the best split
        best_split = {}
        
        # Dùng phương pháp reduction variance để tìm cách chia tốt nhất (giảm dần phương sai -> tìm feature làm độ giảm phương sai lớn nhất)
        max_ig = -float("inf")
        # Lặp qua tất cả các features
        for feature_index in range(num_features):
            feature_values = dataset[:, feature_index]
            
            # Tìm tất cả các ngưỡng có thể
            possible_thresholds = np.unique(feature_values)
            # Lặp qua tất cả các ngưỡng
            for threshold in possible_thresholds:
                # get current split
                dataset_left, dataset_right = self.split(dataset, feature_index, threshold)
                # check if childs are not null
                if len(dataset_left)>0 and len(dataset_right)>0:
                    y, left_y, right_y = dataset[:, -1], dataset_left[:, -1], dataset_right[:, -1]
                    # compute information gain
                    curr_ig = self.information_gain(y, left_y, right_y)
                    # update the best split if needed
                    if curr_ig > max_ig:
                        best_split["feature_index"] = feature_index
                        best_split["threshold"] = threshold
                        best_split["dataset_left"] = dataset_left
                        best_split["dataset_right"] = dataset_right
                        best_split["var_red"] = curr_ig
                        max_ig = curr_ig
                        
        # return best split
        return best_split
    
    def split(self, dataset, feature_index, threshold):
        ''' function to split the data '''
        
        dataset_left = np.array([row for row in dataset if row[feature_index] <= threshold])
        dataset_right = np.array([row for row in dataset if row[feature_index] > threshold])
        return dataset_left, dataset_right
    
    def information_gain(self, parent, left_child, right_child):
        ''' Tính IG = độ giảm phương sai '''
        num_left = len(left_child) / len(parent)
        num_right = len(right_child) / len(parent)

        gain = np.var(parent) - (num_left * np.var(left_child) + num_right * np.var(right_child))
        return gain
                
    def show_tree(self, tree=None, indent=" "):
        ''' function to print the tree '''
        
        if not tree:
            tree = self.root

        if tree.value is not None:
            print(tree.value)

        else:
            print("X_"+str(tree.feature_index), "<=", tree.threshold)
            print("%sTrái:" % (indent), end="")
            self.show_tree(tree.left, indent + " ")
            print("%sPhải:" % (indent), end="")
            self.show_tree(tree.right, indent + " ")
    
    def fit(self, X, Y):
        ''' function to train the tree '''
        
        dataset = np.concatenate((X, Y), axis=1)
        self.root = self.make_tree(dataset)
        
    def make_prediction(self, x, tree):
        ''' function to predict new dataset '''
        
        if tree.value!=None: return tree.value
        feature_val = x[tree.feature_index]
        if feature_val<=tree.threshold:
            return self.make_prediction(x, tree.left)
        else:
            return self.make_prediction(x, tree.right)
    
    def predict(self, X):
        ''' function to predict a single data point '''
        
        preditions = [self.make_prediction(x, self.root) for x in X]
        return preditions
```

## Fit the model

```python
regressor = DecisionTreeRegressor(max_depth=12)
regressor.fit(X_train,y_train)
regressor.show_tree()
```

## Test the model

```python
y_pred = regressor.predict(X_test) 

np.sqrt(mean_squared_error(y_test, y_pred))
```
